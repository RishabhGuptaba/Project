```python
# we have to convert all charatcter varaible in to numeric categorical variable i.e 1,2,3..... 
# Because scikit learn don't accpet charatcter format.
# scikit learn onlyu accpet only numeric format.
from sklearn import preprocessing

for x in train.columns:
    if train[x].dtype == 'object':
        lbl = preprocessing.LabelEncoder()
        lbl.fit(list(train[x].values))
        train[x] = lbl.transform(list(train[x].values))
```

# Random Forest Model

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
y = train['target']
```

```python
del train['target']
```

```python
X= train
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=1,stratify=y)
```

# Train the Random Forest Classifier  

```python
RDCLF= RandomForestClassifier(n_estimators=500,max_depth=6)
```

```python
RDCLF
```

```python
RDCLF.fit(X_train,y_train)
```

```python
RDCLF.predict(X_test)
```

```python
#Prediction and model accuracy 

prediction = RDCLF.predict(X_test)
print(prediction)
```

```python
acc = accuracy_score(np.array(y_test),prediction)
print(acc)
```

```python
print("The Random Forest Model accuracy is {}".format(acc))
```
